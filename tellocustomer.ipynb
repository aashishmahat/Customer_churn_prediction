```python
x=df.iloc[:,:-1]
y=df['Churn_Yes']
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_scaled=ss.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test ,y_train,y_test=train_test_split(X_scaled,y,test_size=0.25,random_state=42)
```

```python
df['TotalCharges'].isnull().sum()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)*100
```

```python
knn.score(x_train,y_train)*100
```

```python
#overfitting train>test so check nenighbours 
for i in range(1,30):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    print(i,"Train:", knn.score(x_train, y_train),"Test:", knn.score(x_test, y_test))
    
```

```python
#best is 27 neighbors 
knn = KNeighborsClassifier(n_neighbors=27)
knn.fit(x_train, y_train)
print("Final Score:", knn.score(x_test, y_test)*100)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
y_pred=knn.predict(x_test)
cf=confusion_matrix(y_test,y_pred)
print(cf,"precision:",precision_score(y_test,y_pred),"recall:",recall_score(y_test,y_pred),"f1 score:",f1_score(y_test,y_pred))
```

```python
import seaborn as sns
sns.heatmap(cf,annot=True,fmt='d')#fmt d show exact number 
plt.show()
```

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear')#sabai vanda ramro kernel linear nai xa poly ,sigmoid vanda
svm.fit(x_train,y_train)
print("Train:", svm.score(x_train, y_train),"Test:", svm.score(x_test, y_test))
    
```

```python
y_pred_svm=svm.predict(x_test)
cf_svm=confusion_matrix(y_test,y_pred_svm)
print(cf_svm,"precision:",precision_score(y_test,y_pred_svm),"recall:",recall_score(y_test,y_pred_svm),"f1 score:",f1_score(y_test,y_pred_svm))
```
